#Importing the Libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#Importing Dataset

```
dataset = pd.read_csv('Amazon Review1.csv', quoting=3, error_bad_lines=False, encoding= 'unicode_escape')
```

```
print(len(dataset))
# for i in range(0,5000):
#   print(dataset['reviews.text'][i])
```

#Refining the Data and Reviews


```
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpos =[]
for i in range(0,len(dataset)):
  ps = PorterStemmer()
  review = re.sub('[^a-zA-Z]',' ', dataset['reviews.text'][i])
  review = review.lower()
  review = review.split()
  all_stopwords = set(stopwords.words('english'))
  all_stopwords.remove('not')
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpos.append(review)
print(corpos)  
```

```
y = dataset.iloc[:,-1].values
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

```
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
x = cv.fit_transform(corpos).toarray()
```

#Splitting the datasets in train and test

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

#Training the classification model on the train set

```
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(x_train, y_train)
```

#Checking the accuracy of the Model

```
y_pred = classifier.predict(x_test)
from sklearn.metrics import accuracy_score, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
